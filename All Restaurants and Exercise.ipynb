```python
#import dependencies
import pandas as pd
import os
from sqlalchemy import create_engine
from config import postgres_pwd
```

```python
#pull all CSV files for restaurants

#Panera
panera_df = pd.read_csv("Panera/Panera_Nutrition_Facts.csv")
#Jimmy Johns
jj_df = pd.read_csv("JimmyJohns/Jimmy_Johns_Nutrition.csv")
#Quiznos
quiznos_df = pd.read_csv("Quiznos/Quiznos CSVs/quiznos_final.csv")
#subway
subway_df = pd.read_csv("Subway/subway_file.csv")
```

```python
quiznos_df = quiznos_df.drop(columns = ['Unnamed: 0'])
quiznos_df
```

```python
subway_df.columns
```

```python
subway_df.head()
```

```python
subway_df = subway_df.drop(columns = ['Unnamed: 0', 'sandwich_id'])
subway_df['restaurant_name'] = 'Subway'
subway_df = subway_df.drop(columns = ['sandwich_type'])
subway_df['calories_from_fat'] = ''
subway_df = subway_df.rename(columns = {'cholestrol': 'cholesterol'})
subway_df_reordered = subway_df[['sandwich_name', 'sandwich_description', 'restaurant_name',
       'bread_type', 'sandwich_length', 'calories', 'calories_from_fat',
       'total_fat', 'saturated_fat', 'trans_fat', 'cholesterol', 'sodium',
       'total_carbohydrates', 'dietary_fiber', 'sugars', 'protein']]
subway_df_reordered
```

```python
#combine all csv files
starter_df = jj_df.append(quiznos_df)
starter_2_df = starter_df.append(panera_df)
all_sandwiches_df = starter_2_df.append(subway_df)
all_sandwiches_df
```

```python
all_sandwiches_df = all_sandwiches_df.reset_index(drop = True)
all_sandwiches_df
```

```python
all_sandwiches_df.columns
```

```python
all_sandwiches_df = all_sandwiches_df.rename(columns = {'calories': 'calories_cal_',
                                    'calories_from_fat': 'calories_from_fat_cal_',
                                    'total_fat': 'total_fat_grams_',
                                    'saturated_fat': 'saturated_fat_grams_' ,
                                    'trans_fat': 'trans_fat_grams_',
                                    'cholesterol': 'cholesterol_mg_',
                                    'sodium': 'sodium_mg_',
                                    'total_carbohydrates': 'total_carbohydrates_grams_',
                                    'dietary_fiber': 'dietary_fiber_grams_',
                                    'sugars': 'sugar_grams_',
                                    'protein': 'protein_grams_'})
all_sandwiches_df = all_sandwiches_df.fillna("na")
all_sandwiches_df = all_sandwiches_df.replace("","na")
all_sandwiches_df
```

```python
all_sandwiches_df = all_sandwiches_df.replace(',','', regex=True)

all_sandwiches_df = all_sandwiches_df.replace("na","0")

all_sandwiches_df
```

```python
#exporting final file to CSV
all_sandwiches_df.to_csv("final_sandwich_data.csv")
```

```python
#establish connection
rds_connection_string = f"postgres:{postgres_pwd}@localhost:5432/for_now"
engine = create_engine(f'postgresql://{rds_connection_string}')
conn = engine.connect()
trans = conn.begin()

sandwich_creation_sql = f"""
-- Drop table if exists
DROP TABLE IF EXISTS sandwiches;

-- Create the sandwiches table
CREATE TABLE sandwiches (
    sandwich_id SERIAL PRIMARY KEY,
    restaurant_name VARCHAR,
    sandwich_name VARCHAR,
    bread_type VARCHAR,
    sandwich_length VARCHAR,
    sandwich_description VARCHAR,
    calories_cal_ INT,
    calories_from_fat_cal_ INT,
    total_fat_grams_ INT,
    protein_grams_ INT,
    saturated_fat_grams_ INT,
    trans_fat_grams_ INT,
    cholesterol_mg_ INT,
    sodium_mg_ INT,
    total_carbohydrates_grams_ INT,
    dietary_fiber_grams_ INT,
    sugars_grams_ INT
);
"""
#conn.execute("commit")
conn.execute(sandwich_creation_sql)
trans.commit()
conn.close()
```

```python
all_sandwiches_df.to_sql(name='sandwiches', con=engine, if_exists='append', index=False)
```
