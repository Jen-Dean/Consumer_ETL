```python
#import dependencies

import pandas as pd
import os
```

```python
#import CSVs from 'Quiznos CSV' folder

quiznos_1 = pd.read_csv("Quiznos CSVs/Quiznos_1.csv")
quiznos_2 = pd.read_csv("Quiznos CSVs/Quiznos_2.csv")
quiznos_3 = pd.read_csv("Quiznos CSVs/Quiznos_3.csv")

#quiznos_1 has 22 rows

#quiznos_2 has 30 rows

#quiznos_3 has 68 rows
```

```python
#merging all dataframes using .append,
quiznos_df_starter = quiznos_1.append(quiznos_2)
#quiznos_df_starter has 52 rows, checks out
quiznos_df_all = quiznos_df_starter.append(quiznos_3)

quiznos_df_all
```

```python
#find the names of the columns
quiznos_df_all.columns
```

```python
#renaming the columns to match the final postgres table
quiznos_df_renamed = quiznos_df_all.rename(columns = {'Unnamed: 0': 'sandwich_length', 
                                 'Name of Sandwich': 'sandwich_name', 
                                 'Calories (kcal)': 'calories',
                                 'Fat Calories (kcal)' : 'calories_from_fat', 
                                 'Fat (g)': 'total_fat', 
                                 'Saturated Fat (g)' : 'saturated_fat',
                                 'Total Trans Fat (g)' : 'trans_fat', 
                                 'Cholesterol (mg)' : 'cholesterol', 
                                 'Sodium (mg)': 'sodium', 
                                 'Carbs (g)': 'total_carbohydrates', 
                                 'Dietary Fiber (g)' : 'dietary_fiber', 
                                 'Sugars (g)' : 'sugars', 
                                 'Protein (g)' : 'protein'})

quiznos_df_renamed.head()
```

```python
#adding additional columns to df
quiznos_df_renamed['restaurant_name'] = ""
quiznos_df_renamed['sandwich_description'] = ""
quiznos_df_renamed['bread_type'] = ""

quiznos_df_renamed.head()
```

```python
#finding column names and reordering
quiznos_df_reordered = quiznos_df_renamed[['restaurant_name', 'sandwich_name', 'sandwich_description', 
                                          'sandwich_length', 'bread_type', 'calories',
                                          'calories_from_fat','total_fat', 'saturated_fat',
                                          'trans_fat', 'cholesterol', 'sodium', 'total_carbohydrates',
                                          'dietary_fiber', 'sugars', 'protein']]
quiznos_df_reordered.head()
```

```python
#adding values to blank columns
quiznos_df_reordered['restaurant_name'] = 'Quiznos'
quiznos_df_reordered['bread_type'] = 'Bread is a white baguette unless otherwise noted in the description'

quiznos_df_reordered.head()
```

```python
#splitting information

new = quiznos_df_reordered["sandwich_name"].str.split("-", n = 1, expand = True)
new['restaurant_name'] = 'Quiznos'
new
```

```python
new = new.rename(columns = {0: 'sandwich_name',
                     1: 'sandwich_description'})
new
```

```python
#drop sandwich name and description column\
quiznos_df_reordered = quiznos_df_reordered.drop(columns = ['sandwich_name', 'sandwich_description'])

quiznos_df_reordered
```

```python
#add previous splits to original df, reorder

quiznos_df_reordered['sandwich_name'] = new['sandwich_name']
quiznos_df_reordered['sandwich_description'] = new['sandwich_description']
quiznos_df_final = quiznos_df_reordered[['restaurant_name','sandwich_name', 'sandwich_description', 'sandwich_length', 'bread_type', 'calories',
       'calories_from_fat', 'total_fat', 'saturated_fat', 'trans_fat',
       'cholesterol', 'sodium', 'total_carbohydrates', 'dietary_fiber',
       'sugars', 'protein']]

quiznos_df_final
```

```python
#reset index for more meaningful data
quiznos_df_final = quiznos_df_final.reset_index(drop=True)
quiznos_df_final
```

```python
#drop rows were value in 'sandwich length' is 'Half Salad' or 'Full Salad'
quiznos_df_final['sandwich_length'].value_counts()
#There are 19 'Half Salad' and 17 'Full Salad' values
quiznos_df_final.drop(quiznos_df_final.loc[quiznos_df_final['sandwich_length']== 'Half Salad'].index, inplace=True)
quiznos_df_final.drop(quiznos_df_final.loc[quiznos_df_final['sandwich_length']== 'Full Salad'].index, inplace=True)
quiznos_df_final
```

```python
#reindex after dropping the salads
quiznos_df_final = quiznos_df_final.reset_index(drop=True)
quiznos_df_final
```

```python
#export final df as csv for Loading process, completion of transformation of Data
quiznos_df_final.to_csv('Quiznos CSVs/quiznos_final.csv')
```
