```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
def init_browser():
    executable_path = {"executable_path": ChromeDriverManager().install()}
    return Browser("chrome", **executable_path, headless=False)
```

```python
browser = init_browser()
listings = {}

url = "https://www.nutristrategy.com/caloriesburned.htm"
browser.visit(url)

html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
table = soup.find_all("table")[3]
table
```

```python
#NO HEADERS

exercises_list = []

for tr in table.find_all('tr')[1:]:
    try:
        exercise1 = tr.find_all('td')[0].get_text().split("\n")[0].strip()
        exercise2 = tr.find_all('td')[0].get_text().split("\n")[1].strip()
        calories_130lbs = tr.find_all('td')[1].get_text().strip()
        calories_155lbs = tr.find_all('td')[2].get_text().strip()
        calories_180lbs = tr.find_all('td')[3].get_text().strip()
        calories_205lbs = tr.find_all('td')[4].get_text().strip()
        final_string = f"{exercise1}{exercise2}"
        exercises_list.append(
        {"exercise (1 Hour)":final_string,
         "130 lbs": calories_130lbs,
         "155 lbs": calories_155lbs,
         "180 lbs": calories_180lbs,
         "205 lbs": calories_205lbs
        })
    except IndexError:
        pass
    
exercises_list
```

```python
exercises_df = pd.DataFrame(exercises_list)
exercises_df
```

```python
exercises_df["130 lbs"] = pd.to_numeric(exercises_df["130 lbs"])
exercises_df["155 lbs"] = pd.to_numeric(exercises_df["155 lbs"])
exercises_df["180 lbs"] = pd.to_numeric(exercises_df["180 lbs"])
exercises_df["205 lbs"] = pd.to_numeric(exercises_df["205 lbs"])


exercises_df
```

```python
exercises_df.to_csv("exercise.csv", index=False)
```

```python
per_min_exercises_df = exercises_df
```

```python
per_min_exercises_df["130 lbs"] /= 60
per_min_exercises_df["155 lbs"] /= 60 
per_min_exercises_df["180 lbs"] /= 60 
per_min_exercises_df["205 lbs"] /= 60 
```

```python
per_min_exercises_df = per_min_exercises_df.round(1)
per_min_exercises_df = per_min_exercises_df.rename(columns={"exercise (1 Hour)":"exercise (calories burned per min)"})
per_min_exercises_df


```

```python
per_min_exercises_df.to_csv("per_min_exercise.csv", index=False)
```
