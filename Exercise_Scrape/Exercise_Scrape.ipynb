```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
def init_browser():
    executable_path = {"executable_path": ChromeDriverManager().install()}
    return Browser("chrome", **executable_path, headless=False)
```

```python
browser = init_browser()
listings = {}

url = "https://www.nutristrategy.com/caloriesburned.htm"
browser.visit(url)

html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
table = soup.find_all("table")[3]
table
```

```python
#Exercises without Header

exercises_list = []

for tr in table.find_all('tr')[1:]:
    try:
        exercise1 = tr.find_all('td')[0].get_text().split("\n")[0].strip()
        exercise2 = tr.find_all('td')[0].get_text().split("\n")[1].strip()
        final_string = f"{exercise1}{exercise2}"
        exercises_list.append(
        {"exercise":final_string})
    except IndexError:
        pass
    
exercises_list
```

```python
exercises_df = pd.DataFrame(exercises_list)
exercises_df
```
